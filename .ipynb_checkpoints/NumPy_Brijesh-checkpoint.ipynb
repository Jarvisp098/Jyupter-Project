```python
import numpy as np
import matplotlib.pyplot as plt

#set bountries for random quantities
min_quantity = 50
max_quantity = 300

np.random.seed(40)

#generate random quantities for week 1 & week 2
week_1_quntities = np.random.randint(min_quantity, max_quantity, size=3)
week_2_quntities = week_1_quntities + np.random.randint(-20, 20, size=3)

#product names for the labels
products = ['T-SHirts', 'Jeans', 'Hats']

week_1_array = np.array(week_1_quntities) #array for week 1
week_2_array = np.array(week_2_quntities) #array for week 2

#Print the properties of the arrays
print("Week 1 quntities:-", week_1_array)
print("Shape for week 1:- ", week_1_array.shape)

print("Week 2 quntities:-", week_2_array)
print("Shape for week 2:- ", week_2_array.shape)

#Calculate the chnage in quntities
qunantity_chnage = week_2_array - week_1_array

print("\nChange in quntities:-", qunantity_chnage)

#Price for the products
prices = np.array([10, 15, 20]) #Price for T-shirt, Jeans and Hats
inventory_value_change = np.dot(qunantity_chnage, prices)

print("\nInventory value change:-", inventory_value_change)

#Concatenate the data into a 2d array
inventory_data = np.concatenate(([week_1_array], [week_2_array]))
print("\nInventory data:-", inventory_data)
print("Shape of inventory data:-", inventory_data.shape)
print("Number of dimensions:-", inventory_data.ndim)

#Plotting
plt.plot(products, week_1_array, marker = 'o', label='Week 1', color='blue')
plt.plot(products, week_2_array, marker = 'o', label='Week 2', color='Orange')

#Adding labels and title
plt.xlabel('Products')
plt.ylabel('Quntities')
plt.title('Inventory levels:- Week 1 vs Week 2')
plt.legend()

#show plot
plt.grid(True)
plt.show()

```
